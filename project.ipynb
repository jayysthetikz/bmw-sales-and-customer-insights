```python
df.columns = df.columns.str.strip().str.replace(' ','_').str.lower()
```

```python
df.columns.tolist()
```

```python
df = df.drop(columns='not_useful_column')
```

```python
df
```

```python
df.last_name = df.last_name.str.strip('/..._').fillna('-')
```

```python
df[['paying_customer', 'do_not_contact']] = df[['paying_customer', 'do_not_contact']].replace('Yes', 'Y').replace('No', 'N').fillna('-').replace('N/a', '-')
df.address = df.address.replace('N/a', '-')
```

```python
df
```

```python
df.phone_number = df.phone_number.apply(lambda x: str(x))
```

```python
df.phone_number = df.phone_number.fillna('-').str.replace('/', '', regex = False).str.replace('|', '', regex = False).str.replace('-', '', regex = False).fillna('-').replace('Na', '-').replace('nan', '-')
```

```python
df1 = pd.read_csv('bmw.csv')
```

```python
df1.head()
```

```python
# Average Price By Transmission
# Price with respect to Mileae and Fuel-Type
# Price Distribution ay boxplot
# Price Distribution with kde

figure , axes = plt.subplots(ncols=2, nrows=2, figsize = (16, 12))

sns.barplot(data=df1, x = 'transmission', y = 'price', ax = axes[0,0], color = 'purple', alpha = 0.2)
axes[0,0].set_title('Transmission V/S Price Plot ~')

sns.scatterplot(data=df1, x = 'price', y = 'mileage', hue = 'fuelType', ax = axes[0, 1], palette='pastel', alpha = 0.5)
axes[0,1].set_title('Price wrt Mileage and Fuel-Type ~')

sns.histplot(data=df1, x = 'price', ax = axes[1,0], bins = 30, color = 'purple', alpha = 0.5, kde=True)
axes[1,0].set_title('Distribution of Prices ~')

sns.boxplot(data=df1, x = 'price', ax = axes[1,1], palette='pastel')
axes[1,1].set_title('Box Plot for Price Distribution ~')

plt.show()
```

```python
# Tax and Fuel-Type Realtion 

plt.figure(figsize=(12, 6))
sns.barplot(data=df1, x = 'fuelType', y = 'tax', palette='pastel')
plt.title('Tax V/S Fuel-Type Relation ~')
plt.show()
```

```python
# Price Trends by Years

plt.figure(figsize = (10, 4))
sns.lineplot(data = df1, x = 'year', y = 'price')
plt.grid(True, alpha = 0.1)
plt.title('Price Trends')
plt.show()
```
